```python
def first_info(dataframe, data_title='Неопознанный набор данных',
               head_length=6, bins=42, figsize=(15,12),
               head_show=True, tail_show=False,
               info_show=True, describe_show=True, corr_show=True, hist_show=True):
    '''
    dataframe: датафрейм для анализа
    data_title: пользовательское название набора данных
    head_length: количество строк, отображаемых функцией при выводе датасета
    bins: для гистограммы - количество корзин, 42 by default
    figsize: для гистограммы - размер графика, кортеж (15,12) by default
    info_show: True, если нужно показывать результат метода DataFrame.info(), True by default 
    describe_show: True, если нужно показывать результат метода DataFrame.describe(), True by default 
    corr_show: True, если нужно показывать результат метода DataFrame.сorr(), True by default 
    hist_show: True, если нужно показывать гистограмму распределений значений DataFrame, True by default
    
    Функция  принимает на вход Pandas.DataFrame, Краткое название датафрейма (для вывода в логе выполнения функции) 
    и количество строк датафрейма, которое будет выводить функция в ходе работы. 
    Функция создана для упрощения первичного знакомства с данными
    Возвращает None
    '''
    try:
        print(f'Данные по датасету: {data_title}')
        if head_show:
            print(f'Первые {head_length} объектов датасета')
            display(dataframe.head(head_length))
        if tail_show:
            print(f'Последние {head_length} объектов датасета')
            display(dataframe.tail(head_length))
        if info_show:
            dataframe.info()
        if describe_show:
            print()
            print(f'Описательная статистика датасета:')
            display(dataframe.describe())
        if corr_show:
            print('Корреляция значений признаков')
            display(pd.DataFrame(dataframe.corr()))
        if hist_show:
            print()
            print('Гистограммы распределения значений признаков:')
            dataframe.hist(bins=bins, figsize=figsize);
            
        
    except:
        print(f'Ошибка в передаваемых аргументах\nПроверьте ввод!')
```

```python
def decomposed_plot(DecomposeResult, interval=None, figsize=(15, 10), 
                    trend_show=True, seasonality_show=True, resid_show=True):
    '''
    DecomposeResult: Results class for seasonal decompositions
    interval: list-like with the start and end value of the datestamp interval 
              If None then the entire time period is used. None by default
    figsize: tuple, figure width and height in inches, (15, 10) by default 
    trend_show, seasonality_show, resid_show: Boolean, if True then shows the specified graph of the component DecomposeResult
    
    The function shows graphs of one or more components of the DecomposeResult class for the received datestamp interval
    Return None          

    '''
    interval_begin = None
    interval_end = None
    
    if interval != None:
        interval_begin = interval[0]
        interval_end = interval[1]
        
    plt.figure(figsize=figsize)
        
    if trend_show:
        plt.subplot(311)
        decomposed.trend[interval_begin:interval_end].plot(ax=plt.gca(), title = 'Trend', xlabel='')

    if seasonality_show:    
        plt.subplot(312)
        decomposed.seasonal[interval_begin:interval_end].plot(ax=plt.gca(), title = 'Seasonality', xlabel='')
        
    if resid_show:   
        plt.subplot(313)
        decomposed.resid[interval_begin:interval_end].plot(ax=plt.gca(), title = 'Residuals', xlabel='')
        
    plt.tight_layout()
    
    return None
```

```python
def lr_research(X, y, use_pipe_scaler=True, scoring='neg_root_mean_squared_error', cv=None, verbose=0, log_show=False):
    '''
    X: Data to predict on (training data)
    y: Targets (training data)
    use_pipe_scaler: boolean, if True the function also scales features with StandartScaler using Pipeline.
    scoring: string, use all possible values for regression from sklearn, 'neg_root_mean_squared_error' by default
    cv: int, cross-validation generator or an iterable, default=None (to use the default 5-fold cross validation)
    verbose: int, The verbosity level, 0 by default
    log_show: True if the operation log and metric values should be displayed while the function is running. False by default
    
    Function creates and trains a model (sklearn LinearRegression) on the training data using a cross-validation metod.
    The quality of the prediction is estimated by the metric RMSE.
    Returns the fitted model and value of RMSE metrics on the data.
    '''
    add_title = ''
    
    if use_pipe_scaler:
        model = Pipeline([("scaler", StandardScaler()), ("LinearRegression", LinearRegression())])
        add_title = '\nС МАСШТАБИРОВАНИЕМ признаков'
    else:
        model = LinearRegression()
                
    if log_show:
        print('Исследование модели ЛИНЕЙНАЯ РЕГРЕССИЯ,',
              add_title,
              '\nкросс-валидация средствами sklearn:', 
              '\nМетрика: ', scoring, end='')
        print()

    scores_lr_auto_cross_val = cross_val_score(model,
                                               X, y,
                                               scoring=scoring,
                                               cv=cv,
                                               verbose=verbose)
    
    train_best_score_lr = (abs(scores_lr_auto_cross_val)).mean()
        
    if log_show:
        print(f'Лучшее значение метрики на обучающем наборе = {train_best_score_lr:.2f}')
        
    return model, train_best_score_lr
```

### Анализ стационарности временного ряда

Оценка стационарности временного ряда - важная часть анализа. Стационарные ряды легче прогнозировать, их характеристики постоянны во времени

Судя по скользящей средней ряд не стационарен. Однако степень стационарности (не стационарности) определим количественно, тестом Дики-Фуллера.

Тест Дики-Фуллера является одним из тестов на единичные корни. Был предложен в 1979 году Дэвидом Дики и Уэйном Фуллером.

Основная идея теста заключается в подтверждении или отклонении двух гипотез:

Нулевая гипотеза (H0): предполагает, что временной ряд имеет единичный корень, то есть он нестационарный.

Альтернативная гипотеза (H1): предполагает, что временной ряд не имеет единичного корня, то есть он является стационарным.

Выберем два порога уровня значимости для теста: 1% и 5%, и сравним полученные результаты

Если p–values больше значения порога, то мы отвергаем нулевую гипотезу, и, следовательно, ряд является стационарным, имеет определенную периодичность, которую возможно выявить.

```python
df_test = adfuller(data['num_orders'])
print('adf', df_test[0])
print('p-value', df_test[1])
print('critical values', df_test[4])

 
print()    
print('Проверка по уровню значимости 5%')
if df_test[0] > df_test[4]['5%']:
    print('Есть единичные корни, ряд не стационарен')
else:
    print('Единичных корней нет, ряд стационарен')
    
print()    
print('Проверка по уровню значимости 1%')
if df_test[0] > df_test[4]['1%']:
    print('Есть единичные корни, ряд не стационарен')
else:
    print('Единичных корней нет, ряд стационарен')
```

Тест Дики - Фулера не дает однозначный ответ о стационарности ряда - при разных уровнях значимости он показывает противоположные результаты.

### Задача прогнозирования

Задача прогнозирования временного ряда состоит в построении модели, которая по историческим данным предскажет будущие значения временного ряда.

Промежуток времени в будущем, на который строится прогноз, называется горизонтом прогнозирования (англ. forecast horizont). В настоящей задаче по условиям он равен одному ЧАСУ. Ранее было выполнено ресеплирование данных с шагом для такого интервала.

Значения временного ряда, или функция x(t), где t — время, — это числа (количество заказов), то есть мы решаем задачу регрессии для временных рядов.

### Создание признаков для прогнозирования

Для задачи обучения модели создадим признаки из исходных данных (временного штампа)

Из вывода по разделу Подготовка выявлено, что данные охватывают период с марта по август 2018 г включительно, поэтому наиболее очевидные признаки, это:
1. Календарные признаки - создаем из исходных данных - значения datetime
 - День
 - Час
 - День недели
2. «Отстающие значения» (англ. lag features) - предыдущие значения временного ряда подскажут, будет ли функция x(t) расти или уменьшаться. 
3. Скользящее среднее - как признак задаёт общий тренд временного ряда.

Объявим функцию для создания указанных признаков

Скользящее среднее и скользящее стандартное отклонение были добавлены в датасет ранее.

```python
def make_features(dataframe, target_name, max_lag):
    '''
    '''
    df = dataframe.copy() # this will make a deep copy of the dataset, to avoid changing the original dataset
    df['day'] = df.index.day
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df[target_name].shift(lag)
        
    return df
```

Добавим признаки. Отстающие значения примем для последнего 72-часового интервала, согласно выводам из раздела 2: Анализ данных

```python
data_proc = make_features(data, target_name='num_orders', max_lag=72)
```
